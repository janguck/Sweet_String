```python
with open("../data/BGCON_CUST_DATA.csv", "rb") as input:  
    data = input.read()
data = data.decode("utf-16")
data = data.encode("utf-8")
data = data.decode("utf-8")
with open("../data/BGCON_CUST_DATA_utf_8.txt", "w") as output:  
    output.write(data)
print(data[:10000])
```
